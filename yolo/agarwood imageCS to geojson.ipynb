```python
for index,row in data.iterrows():
    f_items = ['Agarwood']
    for i in f_items:
        if not os.path.exists(os.path.join(dest_folder, i)):
            os.makedirs(os.path.join(dest_folder, i))
    
    Agarwood = []
    image = row['sourceImage']
    response = row['response']
#     print(image)
    file_name = image.split('/')[-1]
    print file_name
    response = response[1:-1]
#     response = json.loads(response)
#     print (response)
    coordinates=""
    category=""
    response = response.split(":")
#     print(response)
    for cord in range(2,len(response),6):
        coordinates += (response[cord].split(',"')[0])+":::"
        category += response[cord+4].split(',{"')[0].split('}')[0][1:-1]+"&&"
#     print(coordinates)
    print(category)
    for n_cord, n_cat in zip(coordinates.split(':::'), category.split("&&")):
        coordinates = (n_cord[1:-1])
        category = n_cat
#         coords = []
#         coords += [int(a) for a in coordinates.split(",")]
#         print(coordinates)
#         print(nos)
#         print(coordinates)
#         print(category.lower())
        if category.lower().startswith("agar"):
            Agarwood.append(str(coordinates))
            f = open(os.path.join(dest_folder,"Agarwood",file_name+".txt"),"w")
            f.write(str(len(Agarwood))+"\n")
            for idli in Agarwood:
                f.write(str(idli)+"\n")
            f.close()
            
        
            
```
